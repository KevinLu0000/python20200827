# 高雄市公有路外停車場

參考連結：https://data.kcg.gov.tw/dataset/department-of-transportation30/resource/fe3f93da-9673-4f7b-859c-9017d793f798

```python
import pandas as pd
```

```python
url = "https://data.kcg.gov.tw/dataset/449e45d9-dead-4873-95a9-cc34dabbb3af/resource/fe3f93da-9673-4f7b-859c-9017d793f798/download/109.7.1.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
import folium
import geocoder
```

```python
center = geocoder.osm("高雄市").latlng
```

```python
m = folium.Map(location=center, zoom_start=16)
```

```python
m
```

```python
for i, row in df.iterrows():
    try:
        folium.Marker([float(row['緯度']), float(row['經度'])], popup=row['停車場名稱']).add_to(m)
    except Exception as e:
        print(e.args)
```
